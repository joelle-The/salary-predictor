Removing Outliers

```python
cap_gain_threshold = df['capital-gain'].quantile(0.995)
df = df[df['capital-gain'] <= cap_gain_threshold]

cap_gain_cap = df['capital-gain'].quantile(0.99)
df['capital-gain'] = df['capital-gain'].clip(upper=cap_gain_cap)

cap_loss_threshold = df['capital-loss'].quantile(0.995)
df = df[df['capital-loss'] <= cap_loss_threshold]

cap_loss_cap = df['capital-loss'].quantile(0.99)
df['capital-loss'] = df['capital-loss'].clip(upper=cap_loss_cap)

hours_threshold = df['hours-per-week'].quantile(0.995)
df = df[df['hours-per-week'] <= hours_threshold]

hours_cap = df['hours-per-week'].quantile(0.99)
df['hours-per-week'] = df['hours-per-week'].clip(upper=hours_cap)
```

Log transformation is used for capital-gain and capital-loss to reduce skewness, handle extreme outliers, and make distributions more normal-like, improving model performance and stability, especially for linear models.


```python
import numpy as np

df['capital-gain'] = df['capital-gain'].apply(lambda x: np.log1p(x))
df['capital-loss'] = df['capital-loss'].apply(lambda x: np.log1p(x))
```

scaling age, education-num, and hours-per-week because they have different value ranges. Scaling ensures each feature contributes equally, prevents bias toward larger numbers, and improves model performance especially for algorithms sensitive to feature magnitude like logistic regression or SVM.


Scale the variables that need to be scaled to see if scaling or scaling should be used to train the data set. See which which metrics to use

```python
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score, 
    roc_auc_score, average_precision_score, classification_report,
    confusion_matrix, balanced_accuracy_score
)
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import numpy as np

X_train_combined_scaled = X_train_combined.copy()
X_test_combined_scaled = X_test_combined.copy()

cols_to_scale = ['age', 'hours-per-week']

col_indices = [df.columns.get_loc(col) for col in cols_to_scale]

scaler = StandardScaler()
X_train_combined_scaled[:, col_indices] = scaler.fit_transform(X_train_combined[:, col_indices])
X_test_combined_scaled[:, col_indices] = scaler.transform(X_test_combined[:, col_indices])

def evaluate_imbalanced_model(y_true, y_pred, y_pred_proba=None, pos_label=' >50K'):
    print("CLASSIFICATION METRICS")
    
    accuracy = accuracy_score(y_true, y_pred)
    balanced_acc = balanced_accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, pos_label=pos_label)
    recall = recall_score(y_true, y_pred, pos_label=pos_label)
    f1 = f1_score(y_true, y_pred, pos_label=pos_label)
    
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Balanced Accuracy: {balanced_acc:.4f}")
    print(f"Precision (>50K): {precision:.4f}")
    print(f"Recall (>50K): {recall:.4f}")
    print(f"F1 Score (>50K): {f1:.4f}")
    
    if y_pred_proba is not None:
        y_true_binary = np.array(y_true).ravel() == pos_label
        roc_auc = roc_auc_score(y_true_binary, y_pred_proba)
        print(f"ROC AUC: {roc_auc:.4f}")
        
        avg_precision = average_precision_score(y_true_binary, y_pred_proba)
        print(f"Average Precision (PR AUC): {avg_precision:.4f}")
    else:
        roc_auc = None
        avg_precision = None
    
    cm = confusion_matrix(y_true, y_pred)
    print(f"\nConfusion Matrix:")
    print(f"True Negatives (<=50K): {cm[0,0]}")
    print(f"False Positives (<=50K predicted as >50K): {cm[0,1]}")
    print(f"False Negatives (>50K predicted as <=50K): {cm[1,0]}")
    print(f"True Positives (>50K): {cm[1,1]}")
    
    print(f"\nDETAILED CLASSIFICATION REPORT")
    print(classification_report(y_true, y_pred))
    
    return {
        'accuracy': accuracy,
        'balanced_accuracy': balanced_acc,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'roc_auc': roc_auc,
        'avg_precision': avg_precision
    }

print("LOGISTIC REGRESSION (UNSCALED)")
lr_unscaled = LogisticRegression(max_iter=1000)
lr_unscaled.fit(X_train_combined, y_train_combined)
y_pred_lr_unscaled = lr_unscaled.predict(X_test_combined)
y_pred_proba_lr_unscaled = lr_unscaled.predict_proba(X_test_combined)[:, 1]
evaluate_imbalanced_model(y_test_combined, y_pred_lr_unscaled, y_pred_proba_lr_unscaled)

print("LOGISTIC REGRESSION (SCALED)")
lr_scaled = LogisticRegression(max_iter=1000)
lr_scaled.fit(X_train_combined_scaled, y_train_combined)
y_pred_lr_scaled = lr_scaled.predict(X_test_combined_scaled)
y_pred_proba_lr_scaled = lr_scaled.predict_proba(X_test_combined_scaled)[:, 1]
evaluate_imbalanced_model(y_test_combined, y_pred_lr_scaled, y_pred_proba_lr_scaled)

print("K-NEAREST NEIGHBORS (UNSCALED)")
knn_unscaled = KNeighborsClassifier(n_neighbors=5)
knn_unscaled.fit(X_train_combined, y_train_combined)
y_pred_knn_unscaled = knn_unscaled.predict(X_test_combined)
y_pred_proba_knn_unscaled = knn_unscaled.predict_proba(X_test_combined)[:, 1]
evaluate_imbalanced_model(y_test_combined, y_pred_knn_unscaled, y_pred_proba_knn_unscaled)

print("K-NEAREST NEIGHBORS (SCALED)")
knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_combined_scaled, y_train_combined)
y_pred_knn_scaled = knn_scaled.predict(X_test_combined_scaled)
y_pred_proba_knn_scaled = knn_scaled.predict_proba(X_test_combined_scaled)[:, 1]
evaluate_imbalanced_model(y_test_combined, y_pred_knn_scaled, y_pred_proba_knn_scaled)

print("SUPPORT VECTOR MACHINE (UNSCALED)")
svm_unscaled = SVC(probability=True)
svm_unscaled.fit(X_train_combined, y_train_combined)
y_pred_svm_unscaled = svm_unscaled.predict(X_test_combined)
y_pred_proba_svm_unscaled = svm_unscaled.predict_proba(X_test_combined)[:, 1]
evaluate_imbalanced_model(y_test_combined, y_pred_svm_unscaled, y_pred_proba_svm_unscaled)

print("SUPPORT VECTOR MACHINE (SCALED)")
svm_scaled = SVC(probability=True)
svm_scaled.fit(X_train_combined_scaled, y_train_combined)
y_pred_svm_scaled = svm_scaled.predict(X_test_combined_scaled)
y_pred_proba_svm_scaled = svm_scaled.predict_proba(X_test_combined_scaled)[:, 1]
evaluate_imbalanced_model(y_test_combined, y_pred_svm_scaled, y_pred_proba_svm_scaled)
```

Justification for Using Scaled Data

Better Model Results
- K-Nearest Neighbors (KNN) accuracy improved from 0.8400 to 0.8555, and F1-score from 0.6550 to 0.6891 after scaling.
- Logistic Regression accuracy improved from 0.8471 to 0.8477, and ROC AUC increased from 0.8996 to 0.9006 after scaling.
- Support Vector Machine (SVM) ROC AUC improved significantly from 0.8208 to 0.8708, and Average Precision (PR AUC) from 0.6292 to 0.6871 after scaling, although accuracy and F1-score remained unchanged.

Better Convergence and Stability
- Scaling helps algorithms like Logistic Regression and SVM converge more reliably and faster by putting features on t he same scale.

Distance-Based Algorithms Depend on Scale
- Models like K-Nearest Neighbors arehighly sensitive to feature scale because they rely on distance metrics. Scaling ensures that no single feature dominates simply because of its magnitude.

Consistent Feature Importance
- Scaling maintains relative importance of features without bias from differing units or ranges, improving model interpretability.

I decided to use ROC AUC bacause it does not only measure how right the model is but how well the model ranks positive examples higher than the negative ones. ROC AUC evaluates probability ranking and not the final classification threshold. unlike accuracy ROC AUC is not biased by imbalances classes and in my dataset case <=50k takes up more than 3/4 of the target data.
